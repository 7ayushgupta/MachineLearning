# Preprocessing the Data

```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
```

```python
train.head()
```

```python
train
```

```python
np.shape(train)
```

```python
listAgeNA = train.index[pd.isna(train.Age)==True].tolist()
len(listAgeNA)
```

```python
listPclassNA = train.index[pd.isna(train.Pclass)==True].tolist()
len(listPclassNA)
```

```python
listSibSpNA = train.index[pd.isna(train.SibSp)==True].tolist()
len(listSibSpNA)
```

```python
listParchNA = train.index[pd.isna(train.Parch)==True].tolist()
len(listParchNA)
```

```python
listCabinNA = train.index[pd.isna(train.Cabin)==True].tolist()
len(listCabinNA)
```

```python
listEmbNA = train.index[pd.isna(train.Embarked)==True].tolist()
len(listEmbNA)
```

```python
# Deleting the rows not containing the Embarked as it is just 2 in number
train = train.drop(listEmbNA)
np.shape(train)
```

```python
mean_age = np.round(np.mean(train.Age[train.index[pd.isna(train.Age)==False].tolist()]))
print(mean_age)
```

```python
train['Age'][listAgeNA] = mean_age
```

```python
listAgeNA = train.index[pd.isna(train.Age)==True].tolist()
len(listAgeNA)
```

```python
train.Embarked[train.Embarked == 'S'] = 1
train.Embarked[train.Embarked == 'C'] = 3
train.Embarked[train.Embarked == 'Q'] = 3
```

```python
train.Sex[train.Sex == 'male'] = 1
train.Sex[train.Sex == 'female'] = 0
```

```python
train.head()
```

```python
train.Cabin[pd.isna(train.Cabin)==False] = 0
train.Cabin[pd.isna(train.Cabin)] = 1
train.head()
```

```python
Relatives = train.SibSp + train.Parch
```

```python
Relatives[Relatives>0] = 1
train['Alone'] = Relatives
train.head()
```

```python
X = np.asmatrix(train[['Pclass','SibSp','Parch','Sex','Age','Fare','Embarked','Cabin']])
np.shape(X)
```

```python
Y = np.asarray(train.Survived)
np.shape(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_cv,Y_train,Y_cv = train_test_split(X,Y,train_size=0.95,stratify=Y)
```

```python
print(np.shape(X_train))
print(np.shape(X_cv))
print(np.shape(Y_train))
print(np.shape(Y_cv))
```

```python
test = pd.read_csv("test.csv")
test.head()
```

```python
test.Embarked[test.Embarked == 'S'] = 1
test.Embarked[test.Embarked == 'C'] = 3
test.Embarked[test.Embarked == 'Q'] = 3
test.Sex[test.Sex == 'male'] = 1
test.Sex[test.Sex == 'female'] = 0
test.Cabin[pd.isna(test.Cabin)==False] = 0
test.Cabin[pd.isna(test.Cabin)] = 1
Relatives2 = test.SibSp + test.Parch
Relatives2[Relatives2>0] = 1
test['Alone'] = Relatives2
test.head()
```

```python
test = test.fillna(round(test['Age'].mean()))
```

```python
test.head()
```

```python
X_test = np.asmatrix(test[['Pclass','SibSp','Parch','Sex','Age','Fare','Embarked','Cabin']])
np.shape(X_test)
```

# Applying Machine Learning Models to the Preprocessed Data

### 1) SGD Classifier

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd_clf = SGDClassifier()
sgd_clf.fit(X_train,Y_train)
```

```python
Y_pred_sgd = sgd_clf.predict(X_cv)
print(Y_pred_sgd)
```

```python
print("Train accuracy = ",1 - np.mean(np.square(Y_train-sgd_clf.predict(X_train))))
print("Cross Validation accuracy = ",1 - np.mean(np.square(Y_cv-Y_pred_sgd)))
```

```python
test['Survived'] = sgd_clf.predict(X_test)
test[['PassengerId','Survived']].to_csv('SGDSolutions.csv',index=False)
#63.636%
```

### 3) KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k_values = []
error_percent = []

for i in range(1,100):
    classifier = KNeighborsClassifier(n_neighbors=i)
    classifier.fit(X_train,Y_train)
    k_values.append(i)
    error_percent.append(np.mean(np.square(Y_cv-classifier.predict(X_cv))))
    print("n_neighbors = ",k_values[-1]," error = ",error_percent[-1])
    
plt.plot(k_values,error_percent)
```

```python
kn_clf = KNeighborsClassifier(n_neighbors=24)
```

```python
kn_clf.fit(X_train,Y_train)
```

```python
print("Train accuracy = ",1 - np.mean(np.square(Y_train-kn_clf.predict(X_train))))
print("Cross Validation accuracy = ",1 - np.mean(np.square(Y_cv-kn_clf.predict(X_cv))))
```

```python
test['Survived'] = kn_clf.predict(X_test)
test[['PassengerId','Survived']].to_csv('KNNSolutions.csv',index=False)
#66.507%
```
