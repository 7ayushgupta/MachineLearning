```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_table("wine.data", header = None, delimiter = ",")
```

```python
df.head(120)
```

```python
Y = np.asarray(df[0])
print(Y[0:10])
```

```python
del df[0]
df.head()
```

```python
X = np.asmatrix(df)
```

```python
means = [0] 
stds = [1]
for i in range(13):
    means.append(np.mean(X[:,i]))
    stds.append(np.std(X[:,i]))
    X[:,i] = (X[:,i] - means[-1])/stds[-1]
    
print(means)
print(stds)
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
counts = [0,0,0]
for i in Y:
    counts[i-1]+=1
plt.hist(Y,bins=3)
print(counts)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_cv, Y_train, Y_cv = train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=11, stratify =Y)
```

```python
print(np.shape(X_train))
print(np.shape(X_cv))
print(np.shape(Y_train))
print(np.shape(Y_cv))
```

```python
COV = np.round((np.cov(np.transpose(X[:,1:]))),decimals=2)
print(COV)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression(tol=0.0000001,max_iter=1000000,multi_class='multinomial',n_jobs=2,verbose=5,solver='lbfgs')
```

```python
model1.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(model1.predict(X_train),Y_train)
```

```python
accuracy_score(model1.predict(X_cv),Y_cv)
```

```python
confusion_matrix(model1.predict(X_train),Y_train)
```

```python
confusion_matrix(model1.predict(X_cv),Y_cv)
```

```python
X_cv2, X_cv3, Y_cv2, Y_cv3 = train_test_split(X,Y,train_size=0.5,test_size=0.5,random_state=142)
```

```python
print(accuracy_score(model1.predict(X_cv2),Y_cv2))
print(confusion_matrix(model1.predict(X_cv2),Y_cv2))
print(accuracy_score(model1.predict(X_cv3),Y_cv3))
print(confusion_matrix(model1.predict(X_cv3),Y_cv3))
```
